# Missing Values

```python
import polars as pl

dtypes = {
  'mpg': pl.Float64,
  'cylinders': pl.Int64,
  'displacement': pl.Float64,
  'horsepower': pl.Int64,
  'weight': pl.Int64,
  'acceleration': pl.Float64,
  'year': pl.Int64,
  'origin': pl.Int64,
  'name': pl.Utf8
 }

df = pl.read_csv("https://data.heatonresearch.com/data/t81-558/auto-mpg.csv",
                 dtypes=dtypes,null_values=['?','NA'])

df.filter(
    pl.col('horsepower').is_null()
)
```

```python
df = df.with_column(
    pl.col("horsepower").fill_null(
        pl.col("horsepower").median()
    )
)
df.head()

df.filter(
    pl.col('horsepower').is_null()
)
```

# Modify Columns

```python
import polars as pl

dtypes = {
  'mpg': pl.Float64,
  'cylinders': pl.Int64,
  'displacement': pl.Float64,
  'horsepower': pl.Int64,
  'weight': pl.Int64,
  'acceleration': pl.Float64,
  'year': pl.Int64,
  'origin': pl.Int64,
  'name': pl.Utf8
 }

df = pl.read_csv("https://data.heatonresearch.com/data/t81-558/auto-mpg.csv",
                 dtypes=dtypes,null_values=['?','NA'])
df = df.with_column(
    pl.col('weight').apply(lambda x: round(x/2.205))
)
df.head()
```

# Add Columns

```python
import polars as pl

dtypes = {
  'mpg': pl.Float64,
  'cylinders': pl.Int64,
  'displacement': pl.Float64,
  'horsepower': pl.Int64,
  'weight': pl.Int64,
  'acceleration': pl.Float64,
  'year': pl.Int64,
  'origin': pl.Int64,
  'name': pl.Utf8
 }

df = pl.read_csv("https://data.heatonresearch.com/data/t81-558/auto-mpg.csv",
                 dtypes=dtypes,null_values=['?','NA'])
df = df.with_column(
    pl.col('weight').apply(lambda x: round(x/2.205)).alias('weight_kg')
)
df.head()
```

# Group By

```python
import polars as pl

df = pl.read_csv("https://data.heatonresearch.com/data/t81-558/iris.csv")
df.groupby("species").agg(
    [
      pl.col('sepal_l').mean(),
      pl.col('sepal_w').mean(),
      pl.col('petal_l').mean(),
      pl.col('petal_w').mean()
    ]
)
```

# Sort

```python
import polars as pl

dtypes = {
  'mpg': pl.Float64,
  'cylinders': pl.Int64,
  'displacement': pl.Float64,
  'horsepower': pl.Int64,
  'weight': pl.Int64,
  'acceleration': pl.Float64,
  'year': pl.Int64,
  'origin': pl.Int64,
  'name': pl.Utf8
 }

df = pl.read_csv("https://data.heatonresearch.com/data/t81-558/auto-mpg.csv",
                 dtypes=dtypes,null_values=['?','NA'])
df = df.sort("mpg",reverse=True)
df
```

# Lazy

```python
import polars as pl

df = pl.read_csv("https://data.heatonresearch.com/data/t81-558/iris.csv").lazy()
df.groupby("species").agg(
    [
      pl.col('sepal_l').mean(),
      pl.col('sepal_w').mean(),
      pl.col('petal_l').mean(),
      pl.col('petal_w').mean()
    ]
).sort("species")
```

```python
import polars as pl

df = pl.read_csv("https://data.heatonresearch.com/data/t81-558/iris.csv").lazy()
df.groupby("species").agg(
    [
      pl.col('sepal_l').mean(),
      pl.col('sepal_w').mean(),
      pl.col('petal_l').mean(),
      pl.col('petal_w').mean()
    ]
).sort("species").collect()
```
