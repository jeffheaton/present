```python
principal = 240000 # loan amount
apr = 0.05 # loan interest rate
term = 30 * 12 # loan term in months
```

$ P = \frac{r(PV)}{1-(1+r)^{-n}}$

* P = Payment
* PV = Loan amount
* r = Rate per period
* n = number of periods

```python
current = principal
monthly_rate = apr / 12
payment = (monthly_rate * principal) / (1-(1+monthly_rate)**(-term))

print( f"Monthly payment: {payment:,.2f}")

for i in range(term):
  interest_part = (current * apr)/12
  principal_part = payment - interest_part
  print(f"Payment #{i+1}: remaining: ${current:,.2f}, p: ${principal_part:,.2f}, i: ${interest_part:,.2f}")
  current -= principal_part
```
