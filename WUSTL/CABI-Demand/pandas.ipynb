# Pandas Introduction

Begin by loading a dataset about cars.

```python
import pandas as pd

df = pd.read_csv("https://data.heatonresearch.com/data/t81-558/auto-mpg.csv")
df.head()
```

Get all of the cars that are 8 cylinder.

```python
df2 = df[df.cylinders==8]
df2.head()
```

How many cars had 8 cylinders?

```python
len(df2)
```

Display the one car that had 8 cylinders and still had a MPG greater than 25.

```python
df2 = df[ (df.cylinders==8) & (df.mpg>25)]
df2.head()
```

Rename the that car.

```python
df2.loc[364,'name'] = 'Oldsmobile Cutlass LS'
```

Don't modify the sub query df2 (view)!! Always modify the original. 

```python
df.loc[364,'name'] = 'Oldsmobile Cutlass LS'
```

Or combine the query.

```python
q = (df.cylinders==8) & (df.mpg>25)
df.loc[q,'name'] = 'Oldsmobile Cutlass LS'
```

Select only some of the fields.

```python
df[['mpg','cylinders','name']]
```
